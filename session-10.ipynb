```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.exponential(scale=1.0, size=1000)
```

```python
# Create a histogram with 100 bins
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(data, bins=100, edgecolor='black')

# Label the plot
ax.set_title("Histogram of Exponential Distribution Samples")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()

fig.savefig("exponential_histogram.pdf", format="pdf", bbox_inches="tight")
```

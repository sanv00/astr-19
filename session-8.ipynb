```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def f_sin(x):
    return np.sin(x)

def f_cos(x):
    return np.cos(x)
```

```python
# Create a figure with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Left panel: sin(x)
axes[0].plot(x, f_sin(x))
axes[0].set_title("sin(x)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")

# Right panel: cos(x)
axes[1].plot(x, f_cos(x))
axes[1].set_title("cos(x)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")

plt.tight_layout()  # adjusts spacing
plt.show()

fig.savefig("sin_cos_plot.pdf", format="pdf", bbox_inches="tight")
```

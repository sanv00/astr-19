```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.uniform(0, 1, 1000)
```

```python
# Create a histogram with 100 bins
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(data, bins=100, color='skyblue', edgecolor='black')

# Label the axes
ax.set_title("Histogram of Uniform Random Numbers")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()

fig.savefig("uniform_histogram.pdf", format="pdf", bbox_inches="tight")
```
